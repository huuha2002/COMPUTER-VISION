```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
image_path = 'C:/Users/creep/Documents/THIGIACMAYTINH/HOA2.jpg'
image = Image.open(image_path)
```

```python
gray_image = image.convert('L')
plt.imshow(gray_image)
```

```python
# Chuyển ảnh xám về mảng
image_a = np.array(gray_image)
```

```python
# Tính gradient theo hướng x
gradient_x = ndimage.sobel(image_a, axis=1, mode='constant')

# Hiển thị ảnh
fig, axs = plt.subplots(2,1, figsize=(10, 8))

plt.subplot(1,2,1)
plt.imshow(gray_image, cmap='gray')
plt.title('Ảnh xám gốc')

plt.subplot(1,2,2)
plt.imshow(gradient_x, cmap='gray')
plt.title('Ảnh thay doi')

plt.show()
```
