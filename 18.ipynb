```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image
```

```python
image = Image.open('C:/Users/creep/Documents/THIGIACMAYTINH/BANGCHUCAI.jpg')
display(image)
```

```python
# Chuyn sang nh xm
gray_image = image.convert('L')
```

```python
image_array = np.array(gray_image)
```

```python
# Chuyn sang nh nh phn
binary_image = 1*(image_array<200)
```

```python
closing_image = ndimage.binary_closing(binary_image, structure=np.ones((7,7)))
```

```python
# nh nhn
labeled_image, num_features = ndimage.label(closing_image)
```

```python
fig, axs = plt.subplots(3, 1, figsize=(50, 50))

plt.subplot(3,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("nh xm")

plt.subplot(3,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("nh nh phn")

plt.subplot(3,1,3)
plt.imshow(labeled_image, cmap='gray')
plt.title(f" nh nhn ({num_features} i tng)")

plt.show();
```
