```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image
```

```python
image = Image.open('C:/Users/creep/Documents/THIGIACMAYTINH/BANGCHUCAI.jpg')
display(image)
```

```python
# Chuyển sang ảnh xám
gray_image = image.convert('L')
```

```python
image_array = np.array(gray_image)
```

```python
# Chuyển sang ảnh nhị phân
binary_image = 1*(image_array<200)
```

```python
closing_image = ndimage.binary_closing(binary_image, structure=np.ones((7,7)))
```

```python
# Đánh nhãn
labeled_image, num_features = ndimage.label(closing_image)
```

```python
fig, axs = plt.subplots(3, 1, figsize=(50, 50))

plt.subplot(3,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("Ảnh xám")

plt.subplot(3,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("Ảnh nhị phân")

plt.subplot(3,1,3)
plt.imshow(labeled_image, cmap='gray')
plt.title(f"Đã đánh nhãn ({num_features} đối tượng)")

plt.show();
```
